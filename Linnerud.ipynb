```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn import datasets  # importação do datasets
from tabulate import tabulate  # Para exibir tabelas formatadas no console

# Adicionando textos explicativos
texto_intro = """
Antes de explicar o que é o Linnerrud dataset, vamos explicar o que é um dataset.
Os datasets são bases de dados específicas que servem de amostras para treinamentos de algoritmos de inteligência artificial ou para outros tipos de projetos de Data Science.  Os dados de um dataset podem ser usados para treinamento de um algoritmo de machine learning que vai prever alguma informação, como também pode ser base para a visualização de dados com gráficos e relatórios que descrevem a base e extraem insights de forma mais direta.
O conjunto de dados Linnerud é amplamente utilizado em aprendizado de máquina e estatísticas. Ele é uma ferramenta essencial para analisar as relações entre características físicas e o desempenho em atividades físicas. Para entender melhor esse conjunto de dados, é importante conhecer sua estrutura, conteúdo e possíveis aplicações.


Dados -

Peso (Weight): Representa o peso de cada indivíduo no conjunto de dados.
Cintura (Waist): Indica a medida da cintura dos participantes.
Pulso (Pulse): Reflete a taxa de pulso dos indivíduos durante o estudo.
Dados - de atividades físicas realizadas por cada participante

Barra (Chins): Refere-se ao número de elevações na barra fixa que cada participante conseguiu realizar. Esse exercício mede principalmente a força dos músculos dos braços e das costas.
Abdominais (Situps): Representa a quantidade de abdominais que os indivíduos completaram. Esse exercício é utilizado para avaliar a força e resistência dos músculos abdominais.
Saltos (Jumps): Indica o número de saltos verticais realizados por cada participante. Esse exercício é uma medida de força explosiva das pernas e coordenação.

Essas três variáveis foram coletadas para avaliar o desempenho físico dos participantes em diferentes tipos de exercícios que envolvem força, resistência e explosão muscular.

Exemplo de variável quantitativas ( Dados que podem ser medidos numericamente) -

Dentro do Linnerrud dataset possuimos como variáveis quantitativas:
        Peso - Quanto o participante pesa.
        Cintura - Medida da cintura.
        Pulso - Quantos batimentos cardíacos .
        Barra - Número de barras físicas realizadas pelo participante.
        Abdominais - Número de abdominais realizadas pelo participante.
        Saltos -  Número de salto realizadas pelo participante.

Dentro  Linnerrud dataset, não possuímos dados qualitativos, apenas dados númericos.
"""

print(texto_intro)

# Carregar o dataset Linnerud
linnerud_data = datasets.load_linnerud()

# Converter em DataFrame
df = pd.DataFrame(linnerud_data.data, columns=linnerud_data.feature_names)
df_targets = pd.DataFrame(linnerud_data.target, columns=linnerud_data.target_names)

# Exibir as primeiras 5 linhas do dataset
print("\n⤷ Dados dos exercícios:")
print(tabulate(df.head(), headers='keys', tablefmt='psql'))

print("\n⤷ Dados fisiológicos dos participantes:")
print(tabulate(df_targets.head(), headers='keys', tablefmt='psql'))

# Mostrar o tamanho do dataset
print(f"\nTamanho dos dados de exercícios: {df.shape}")
print(f"Tamanho dos dados fisiológicos: {df_targets.shape}")

# Listar as colunas
print("\nColunas dos dados dos exercícios:")
print(tabulate(pd.DataFrame(df.columns, columns=["Colunas dos Dados dos Exercícios"]), headers='keys', tablefmt='psql'))

print("\nColunas dos dados fisiológicos dos participantes:")
print(tabulate(pd.DataFrame(df_targets.columns, columns=["Colunas dos Dados Fisiológicos"]), headers='keys', tablefmt='psql'))

print("\n\n")

# Dados fisiológicos
dados_fisiologicos = {
    'Peso': [191, 189, 193, 162, 189, 182, 211, 167, 176, 154, 169, 166, 154, 247, 193, 202, 176, 157, 156, 138],
    'Cintura': [36, 37, 38, 35, 35, 36, 38, 34, 31, 33, 34, 33, 34, 41, 38, 37, 37, 33, 32, 33],
    'Pulso': [50, 52, 58, 62, 46, 56, 56, 60, 74, 50, 70, 64, 74, 50, 70, 60, 62, 54, 78, 64]
}

# Dados de exercícios
dados_exercicio = {
    'Barras': [5, 2, 12, 12, 13, 4, 8, 6, 15, 17, 17, 13, 14, 1, 6, 12, 4, 11, 15, 2],
    'Abdominais': [162, 110, 101, 105, 155, 101, 50, 31, 120, 60, 80, 82, 125, 40, 95, 100, 150, 73, 110, 50],
    'Saltos': [60, 60, 101, 105, 155, 58, 50, 31, 120, 60, 80, 82, 125, 60, 95, 100, 150, 73, 110, 50]
}

df_fisiologicos = pd.DataFrame(dados_fisiologicos)
df_exercicio = pd.DataFrame(dados_exercicio)

# Calcular as médias
media_fisiologicos = df_fisiologicos.mean()
media_exercicio = df_exercicio.mean()

# Exibir as médias em tabelas
print("\nMédia das Variáveis Fisiológicas:")
print(tabulate(pd.DataFrame(media_fisiologicos, columns=['Média']), headers='keys', tablefmt='psql'))

print("\nMédia das Variáveis de Exercício:")
print(tabulate(pd.DataFrame(media_exercicio, columns=['Média']), headers='keys', tablefmt='psql'))

# Criar gráficos
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].bar(media_fisiologicos.index, media_fisiologicos.values, color='skyblue')
axs[0].set_title('Média das Variáveis Fisiológicas')
axs[0].set_ylabel('Valor Médio')

axs[1].bar(media_exercicio.index, media_exercicio.values, color='lightgreen')
axs[1].set_title('Média das Variáveis de Exercício')
axs[1].set_ylabel('Valor Médio')

plt.tight_layout()
plt.show()

print("\n\n")

# Cálculo das correlações
corr_peso_barras, _ = pearsonr(df_fisiologicos['Peso'], df_exercicio['Barras'])
print(f'Correlação de Pearson entre Peso e Barras: {corr_peso_barras:.2f}')

corr_cintura_abdominais, _ = pearsonr(df_fisiologicos['Cintura'], df_exercicio['Abdominais'])
print(f'Correlação de Pearson entre Cintura e Abdominais: {corr_cintura_abdominais:.2f}')

print("\n\n")

# Exibir correlações em tabelas
correlacoes = pd.DataFrame({
    'Variável 1': ['Peso', 'Cintura'],
    'Variável 2': ['Barras', 'Abdominais'],
    'Correlação': [corr_peso_barras, corr_cintura_abdominais]
})

print(tabulate(correlacoes, headers='keys', tablefmt='psql'))

# Criar gráficos de dispersão
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].scatter(df_fisiologicos['Peso'], df_exercicio['Barras'], color='skyblue')
axs[0].set_title(f'Peso vs Barras (r = {corr_peso_barras:.2f})')
axs[0].set_xlabel('Peso')
axs[0].set_ylabel('Barras')

axs[1].scatter(df_fisiologicos['Cintura'], df_exercicio['Abdominais'], color='lightgreen')
axs[1].set_title(f'Cintura vs Abdominais (r = {corr_cintura_abdominais:.2f})')
axs[1].set_xlabel('Cintura')
axs[1].set_ylabel('Abdominais')

plt.tight_layout()
plt.show()

```
